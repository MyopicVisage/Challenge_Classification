### Build data set

```python
## Method to import Data and Hist Data and generate a training set for model
def importX(filename, filename_hist=0):
    ## Import
    df = read_csv(filename, header=0, delimiter=',', index_col=0, );
    
    ## Clean
    # Boolean
    df['has_gas'].replace(to_replace='t', value=1, inplace=True);
    df['has_gas'].replace(to_replace='f', value=0, inplace=True);
    
    # Date
    df['date_end_ord'] = pd.to_datetime(df['date_end']).map(dt.datetime.toordinal); # 'yyy-mm-dd' to int
    df['date_activ_ord'] = pd.to_datetime(df['date_activ']).map(dt.datetime.toordinal); # 'yyy-mm-dd' to int
    df['date_first_activ_ord'] = pd.to_datetime(df['date_first_activ']).map(dt.datetime.toordinal); # 'yyy-mm-dd' to int
    df['date_modif_prod_ord'] = pd.to_datetime(df['date_modif_prod']).map(dt.datetime.toordinal); # 'yyy-mm-dd' to int
    df['date_renewal_ord'] = pd.to_datetime(df['date_renewal']).map(dt.datetime.toordinal); # 'yyy-mm-dd' to int
    # Drop Original Date columns  
    df.drop(['date_end','date_activ','date_first_activ','date_modif_prod','date_renewal'], axis=1, inplace=True)
    
    # Drop columns missing all data
    def PercentNulls(df):
        percent = df.isnull().sum()/df.count().max()*100;  
        if len(percent[percent > 0])==0:  
            print('Data cleaning complete!')  
            print(percent)
        # else:  
            # print('Percent of column missing data[%]:\n', sep='')  
            # print(percent ) # percent[percent > 0]  # 
        return percent
    percent = PercentNulls(df)
    df.drop(percent[percent>90].index, axis=1, inplace=True)
    
    # Categorical
    # df['activity_new'].fillna('nodata', inplace=True)  
    df['channel_sales'].fillna('nodata', inplace=True)  
    df['origin_up'].fillna('nodata', inplace=True)  
    # Dummies
    # activity_new_dummies = pd.get_dummies(df['activity_new'], prefix='activity_new')  # Doesn't contribute substantial explained var, dropped!   
    channel_sales_dummies = pd.get_dummies(df['channel_sales'], prefix='channel_sales')  
    origin_up_dummies = pd.get_dummies(df['origin_up'], prefix='origin_up')  
    # Drop Originals
    categoricals = ['activity_new', 'channel_sales', 'origin_up'];  
    df.drop(categoricals, axis=1, inplace=True)  
    
    # Concatenate with df  
    # df = pd.concat([df, activity_new_dummies], axis=1, join_axes=[df.index])  
    df = pd.concat([df, channel_sales_dummies], axis=1, join_axes=[df.index])  
    df = pd.concat([df, origin_up_dummies], axis=1, join_axes=[df.index]) 
    
    # Float
    floats = list(df.loc[:, df.dtypes == np.float64].columns)
    
    # Impute missing values from file
    with open('ImputeMean.json') as data_file:  
        impute = json.load(data_file)  
    
    # Replace Null cells in column vector with impute value  
    for i in floats:  
        df[i] = df[i].apply(lambda x: x if not pd.isnull(x) else impute[i] )  # Column command  
    
    ## Handle hist data  
    # Import hist data  
    df2 = read_csv(filename_hist, header=0, delimiter=',', index_col=[0,1], );  
    df2 = pd.DataFrame(df2, dtype='float') # Coerce dtype  
    # Interpolate empty values, where some data exists  
    df2 = df2.fillna(df2.groupby(level=0).transform(lambda x: x.interpolate(method='linear').bfill().ffill() ) )  
    # Fill remaining cells with the mean  
    histImpute = df2.groupby(level=[1]).mean();
    df2 = df2.fillna(histImpute)  
    # Concatenate hist data with main data  
    histPriceAvg = df2.groupby(level=['id'] ).mean();  
    df = pd.concat([df, histPriceAvg], axis=1, join='inner', join_axes=[df.index]);  
    
    # Verify Cleaning Complete  
    PercentNulls(df)  
    
    return df  
```

### Apply Stochastic Gradient Descent Support Vector Machine Classifier

The multi-tool of machine learning, support vector machines draw distictions on cross-sections of the data, and with each sample the classification is refined in accordance with the loss fuction and learning rate. This model is percieved as slower to train than alternative models.

```python
def GridSearchCVSGD(X,y):
    start_time = time.time()  
    
    from sklearn.linear_model import SGDClassifier  
    from sklearn.grid_search import GridSearchCV  
    
    param_grid = {'alpha': (1e-5),  
                  'loss': ('modified_huber',),  
                  'penalty': ('elasticnet',),  
                  'l1_ratio': (0.5),  
                  'n_iter': (50),  
                  'class_weight': ('balanced',)  
                  }  

    import warnings  
    warnings.filterwarnings('ignore');  

    model = SGDClassifier(loss='modified_huber', penalty='elasticnet',
                                          alpha=1e-5, l1_ratio=0.5,
                                          n_iter=50, random_state=42, 
                                          class_weight='balanced', 
                                          shuffle=True, n_jobs=1,
                                          warm_start=False)
    gs_cv = GridSearchCV(model, param_grid, scoring='roc_auc', cv=3, n_jobs=1, verbose=1).fit(X_train, y_train)  
    

    warnings.filterwarnings('default');  

    #Best hyperparameter setting  
    print('Best hyperparameters: %r' % gs_cv.best_params_)  
    print('Best score ({}): {:.4f}'.format(gs_cv.scoring, gs_cv.best_score_)  )

    ## Stochastic Gradient Decent Classifier Model  
    model = SGDClassifier(**gs_cv.best_params_).fit(X_train, y_train)  

    stop_time = time.time();  
    print("Regressor ran in {:.2f}s".format(stop_time - start_time))  
    os.system("set_volume(2)"); os.system("say 'Computation Complete!'");  
    
    return model  
```

```python
## predictAdj: use this method to alter the conservative/liberal bias of the binary classifier
def predictAdj(pred_proba_df, threshold):
    # pred_proba_train_df[pred_proba_train_df>=threshold]  
    predictions = pred_proba_df

    for i in pred_proba_df.index:
        if pred_proba_df.loc[i, 'pred_proba'] >= threshold:
            predictions.loc[i, 'pred_adj'] = 1  
        else:
            predictions.loc[i, 'pred_adj'] = 0  
    return predictions
        
predictions_train = predictAdj(pred_proba_train_df, threshold)

churnersPostAdj = predictions_train.groupby(['pred_adj']).size()[1]; 
print('Post adjustment, number of Churners predicted on training set: {} of {}, or {:.4}'
      .format(churnersPostAdj, y_train.shape[0], churnersPostAdj/y_train.shape[0]) )
predictions_train.head()
```

```python
## Declare plot method to visualize principal components of explained variance
def Importances(clf, X=X_train, n=12):  
    importances = clf.feature_importances_
    std = np.std([clf.feature_importances_],
                 axis=0)
    indices = np.argsort(importances)[::-1]

    # Plot the feature importances of the forest  
    plt.figure(figsize=(9, 6), dpi=100)  
    plt.title("{} Model Variables, Ranked by Feature Importance".format(n))  # Greatest {} Contributors
    plt.bar(range(X.shape[1]), 100*importances[indices],  
           color="r", yerr=std[indices], align="center")  
    plt.xticks(range(X.shape[1]), indices)  
    plt.xlim([-0.5, n+0.5-1])  # X.shape[1]  
    plt.ylim([0, 100*importances.max()*1.10])  
    plt.xlabel('Feature')  
    plt.ylabel('Explained Variance [%]')  
    plt.show()  
    
    # Print the feature ranking  
    print("Feature Ranking: (Explained Variance[%])")  
    for f in range(n):  # X.shape[1]  
        print("%d. Feature %d (%2.2f) %s" % (f + 1, indices[f], importances[indices[f]]*100, X.columns[indices[f]]))  
```
